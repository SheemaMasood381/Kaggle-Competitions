## setting up image generator for test data

```python
# Set up image data generator for test data
test_datagen = ImageDataGenerator(rescale=1./255)

# Load the test images from the extracted directory
test_generator = test_datagen.flow_from_directory(
    directory=extract_path,
    target_size=(32, 32),  # Ensure this matches the input size of your model
    batch_size=1,
    class_mode=None,
    shuffle=False,
    seed=42
)
```

## Make Predictions on test images given for competition on kaggle

```python
predictions = model.predict(test_generator, steps=len(test_generator), verbose=1)
```

```python
# Get the predicted class labels
class_labels = np.argmax(predictions, axis=1)
```
